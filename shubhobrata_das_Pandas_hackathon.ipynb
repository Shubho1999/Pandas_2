**Problem 2**: Write a python function that receives 1 input, batch_name and returns the list of top 10 students who require maximum attention from the teacher.

**Note**: I know this is kind of abstract, but truly we are facing this problem. Try to come up with a solution/algorithm. 

**Tip** : Don't just think in terms of marks, also think about things like late submissions, number of submissions, myday etc

**Tricky scenario**: Aisa bhi ho sakta hai koi student bas shuru me acha perform nai kar raha tha, but ab kar raha hai. So consider dates also.

**Marks : 50**

```
# Write your code here
def improve(id):
  old=data1.merge(data11)
  another=data2.groupby('user_id').sum()
  again=old.merge(another,left_on='user_id',right_on='user_id')
  s=again['batch_id']==id
  cool=again[s]
  x8=data6['submission_date']>='2020-02-01'
  a=data6[x8]
  col1=a.groupby('user_id').count()
  x9=data7['entry_date']>='2020-02-01'
  a2=data7[x9]
  col2=a2.groupby('user_id').count()
  gy=col1['link'].reset_index()
  every1=cool.merge(gy)
  my=col2['entry'].reset_index()
  every2=every1.merge(my)
  x7=data2['evaluation_date']>='2020-02-01'
  a3=data2[x7]
  col3=a3.groupby('user_id').sum()
  scoring=col3['score'].reset_index()
  scoring.rename(columns={'score':'score_after_today'},inplace=True)
  final=every2.merge(scoring)
  x1=final['score_after_today']/2
  y1=final['link']+final['entry']+x1
  calculate=y1*5
  z=final['late']
  z1=10*z
  ultimate=calculate-z1
  final['requires_attention']=ultimate
  ya=final.set_index('requires_attention').sort_index()
  imp=ya.head(10)
  return imp
```

**Problem 3**: Write a python function that will take batch name and a date range as input and outputs the most consistent student in that given daterange and batch

**Note**: Come up with your definition of consistency

**Marks - 20** 

```
# Write your code here
def Consistent(name,daterange):
  old=data1.merge(data11)
  another=data2.groupby('user_id').sum()
  again=old.merge(another,left_on='user_id',right_on='user_id')
  s=again['batch_name']==name
  cool=again[s]
  x8=data6['submission_date']<=daterange
  a=data6[x8]
  col1=a.groupby('user_id').count()
  x9=data7['entry_date']<=daterange
  a2=data7[x9]
  col2=a2.groupby('user_id').count()
  gy=col1['link'].reset_index()
  every1=cool.merge(gy)
  my=col2['entry'].reset_index()
  every2=every1.merge(my)
  x7=data2['evaluation_date']<=daterange
  a3=data2[x7]
  col3=a3.groupby('user_id').sum()
  scoring=col3['score'].reset_index()
  scoring.rename(columns={'score':'score_till_now'},inplace=True)
  final=every2.merge(scoring)
  x3=final['score_till_now']/2
  y1=final['link']+final['entry']+x3
  calculate=y1*5
  z=final['late']
  z1=10*z
  ultimate=calculate-z1
  final['most_consistent']=ultimate
  get=final.set_index('most_consistent').sort_index(ascending=False)
  hmm=get.drop_duplicates('batch_id')
  consistency=hmm['fname'].reset_index()
  return consistency
```
